```python
import cmipld
cmipld.loader.clear_cache()
cmipld.get
```

## Fast load using depth = 0

```python


cmipld.get('vr:variable-root/areacell',depth=0)
```

### To expand substituted items we can increase the depth.

```python
cmipld.get('vr:variable-root/areacell',depth=1)
```

## To extract a specific variable, we key the branded identifier.

We start by seeing the linked files containg data to be substituted.

```python
cmipld.loader.clear_cache()
cmipld.depends('vr:variable/areacell_tavg-u-hxy-u')
```

We can now load the basic file

```python
cmipld.get('vr:variable/areacell_tavg-u-hxy-u',depth=0)
```

or the version with substituted content

```python
cmipld.get('vr:variable/areacell_tavg-u-hxy-u')
```

## Exploring Tables (historic, as current are in progress)
We can load the variables from a CMIP6-style table

```python
cmipld.get('vr:cmip6-mip-table/aerhr',depth=0)
```

Populate the generic variable information

```python
cmipld.get('vr:cmip6-mip-table/aerhr',depth=1)
```

Or list this comprehensively

```python
cmipld.get('vr:cmip6-mip-table/aerhr',depth=4)
```
